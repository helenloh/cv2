```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# Load the pre-trained Haar cascade classifier for license plate detection
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')

def detect_plate(image):
    # Convert the image to grayscale
    gray = image.copy()

    # Detect potential plate rectangles in the image
    plates = plate_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=3)

    # (0, 0, 255) would be pure red.
    # (255, 0, 0) would be pure blue.
    # (255, 255, 255) would be white.
    # (0, 0, 0) would be black.
    # Draw rectangles around the detected plates with green color and thickness to 2
    for (x, y, w, h) in plates:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    return image

# Read an image containing a car
car_image = cv2.imread('../DATA/sport_car.jpg')  # Replace with your car image path
car_image = cv2.cvtColor(car_image, cv2.COLOR_BGR2RGB)
plt.imshow(car_image)


```

```python
# Detect and display the car plate in the image
result_image = detect_plate(car_image)
cv2.imshow('Detected Plate', result_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
